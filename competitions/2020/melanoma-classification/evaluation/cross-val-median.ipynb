Mean across cross-val runs

```python
from os.path import join
from os import listdir
import torch
import pandas as pd
```

```python
configs = [
    'competitions/2020/melanoma-classification/configs/resnet18/cifarbaseline-balanced',
    'competitions/2020/melanoma-classification/configs/resnet18-cv/cifarbaseline-balanced-fold2',
    'competitions/2020/melanoma-classification/configs/resnet18-cv/cifarbaseline-balanced-fold3',
    'competitions/2020/melanoma-classification/configs/resnet18-cv/cifarbaseline-balanced-fold4',
    'competitions/2020/melanoma-classification/configs/resnet18-cv/cifarbaseline-balanced-fold5']
```

```python
all_predictions = None

for index, config in enumerate(configs):
    pred_path = join('/output', config, 'logs/evaluation/test.csv')
    fold_predictions = pd.read_csv(pred_path)
    fold_predictions = fold_predictions.rename(columns={'target': f'target{index+1}'})
    if all_predictions is None:
        all_predictions = fold_predictions
    else:
        all_predictions = pd.merge(all_predictions, fold_predictions, on='image_name')
```

```python
all_predictions
```

```python
target_columns = [f'target{index+1}' for index in range(len(configs))]
```

```python
target_columns
```

```python
all_predictions
```

```python
all_predictions['target'] = all_predictions[target_columns].median(axis=1)
```

```python
all_predictions = all_predictions.drop(columns=target_columns)
```

```python
all_predictions.to_csv('submissions/resnet18-cifarbaseline-balanced-cv-median.csv', index=False)
```
